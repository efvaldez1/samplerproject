```python
df4.show(5)
df1_2 = df1.union(df2)
print((df1_2.count(), len(df1_2.columns)))
df3_4 = df3.union(df4)
print((df3_4.count(), len(df3_4.columns)))
df1_2_3_4=df1_2.union(df3_4)
print((df1_2_3_4.count(), len(df1_2_3_4.columns)))

uniqueIMO=df1_2_3_4.select('mmsi').distinct().collect()
print("Unique mmsi: ",len(uniqueIMO))
#Get data type of a specific column
print(df1_2_3_4.schema["imo"].dataType)
df1_2_3_4 = df1_2_3_4.withColumn("imo", df1_2_3_4["imo"].cast(IntegerType()))
print(df1_2_3_4.schema["imo"].dataType)
df1_2_3_4.printSchema()
df1_2_3_4.orderBy('imo',ascending=True)
df1_2_3_4.createOrReplaceTempView('df')
distinctIMO=spark.sql("""
    SELECT distinct
        imo
    FROM
        df
    """)
print(distinctIMO.show(5))
print(distinctIMO.count())
distinctIMO=distinctIMO.orderBy('imo', ascending=True)
print(distinctIMO.show(15))
```

```python
#read IHS Data
IHS_2020 = spark.read.option("header","true").format("csv").load("IHS/MovementData_2020.CSV")
print((IHS_2020.count(), len(IHS_2020.columns)))
IHS_2019 = spark.read.option("header","true").format("csv").load("IHS/MovementData_2019.CSV")
print((IHS_2019.count(), len(IHS_2019.columns)))
mergedIHS=IHS_2019.union(IHS_2020)
print((mergedIHS.count(), len(mergedIHS.columns)))
mergedIHS.printSchema()
# the only columns that we will be using:
```

```python
print(mergedIHS.show(5))
mergedIHS = mergedIHS.withColumn("LRNOIMOShipNo", mergedIHS["LRNOIMOShipNo"].cast(IntegerType()))
print(type(distinctIMO))
old_columns=distinctIMO.columns
print(old_columns)
new_columns = [column.replace('imo','LRNOIMOShipNo') for column in old_columns]
print(new_columns)
def df_col_rename(X, to_rename, replace_with):
    """
    :param X: spark dataframe
    :param to_rename: list of original names
    :param replace_with: list of new names
    :return: dataframe with updated names
    """
    import pyspark.sql.functions as F
    mapping = dict(zip(to_rename, replace_with))
    X = X.select([F.col(c).alias(mapping.get(c, c)) for c in to_rename])
    return X
distinctIMO=df_col_rename(distinctIMO, old_columns, new_columns)
print("NEW SCHEMA:\n")
distinctIMO.printSchema()
mergedIHS.createOrReplaceTempView('ihs')
print(distinctIMO.count())
IMOlist = distinctIMO.select("LRNOIMOShipNo").rdd.flatMap(lambda x: x).collect()
IMOlist.pop(0)
print(IMOlist[0],IMOlist[1])
print(type(IMOlist))
IMOstring='('+  ','.join([str(i) for i in IMOlist])+')'
# print("IMO STRING: ",IMOstring)
query= """SELECT * FROM ihs WHERE LRNOIMOShipNo IN {}""".format(IMOstring)
# print(query)
merged=spark.sql(query)
print("Number of rows of query result: ", merged.count())
merged=merged.orderBy('LRNOIMOShipNo', ascending=True)
print(merged.show(5))
# q25 = 500
# var2 = 50
# Q1 = spark.sql("SELECT col1 from table where col2>{0} limit {1}".format(var2,q25))
```
